# Student Performance Data Analysis
Using basic pandas operations and visualizations

```python
# 1. Data Loading
import pandas as pd

# Load dataset
df = pd.read_csv("student-mat.csv")

# Display first few rows
df.head()
```

```python
# 2. Data Exploration
# Check for missing values
print("Missing values in each column:")
print(df.isnull().sum())

# Column data types
print("\nData types of columns:")
print(df.dtypes)

# Dataset shape
print("\nShape of the dataset:")
print(df.shape)
```

```python
# 3. Data Cleaning
# Remove duplicate rows
df = df.drop_duplicates()

# Handle missing values (if any)
df = df.fillna(df.median(numeric_only=True))
```

```python
# 4. Data Analysis Questions
# 1. Average G3
average_g3 = df['G3'].mean()
print(f"Average final grade (G3): {average_g3:.2f}")

# 2. Students with G3 > 15
above_15 = df[df['G3'] > 15].shape[0]
print(f"Students scoring above 15 in G3: {above_15}")

# 3. Correlation between studytime and G3
correlation = df['studytime'].corr(df['G3'])
print(f"Correlation between study time and G3: {correlation:.2f}")

# 4. Average G3 by gender
avg_by_gender = df.groupby('sex')['G3'].mean()
print("\nAverage G3 by gender:")
print(avg_by_gender)
```

```python
# 5. Data Visualization
import matplotlib.pyplot as plt
import seaborn as sns

# Histogram of G3
plt.figure(figsize=(8,5))
plt.hist(df['G3'], bins=10, edgecolor='black', color='skyblue')
plt.title("Histogram of Final Grades (G3)")
plt.xlabel("Final Grade (G3)")
plt.ylabel("Number of Students")
plt.grid(True)
plt.show()

# Scatter plot: studytime vs G3
plt.figure(figsize=(8,5))
plt.scatter(df['studytime'], df['G3'], alpha=0.6, color='orange')
plt.title("Study Time vs Final Grade (G3)")
plt.xlabel("Study Time")
plt.ylabel("Final Grade (G3)")
plt.grid(True)
plt.show()

# Bar chart: average G3 by gender
plt.figure(figsize=(6,5))
sns.barplot(data=df, x='sex', y='G3', errorbar=None, palette='pastel')
plt.title("Average Final Grade (G3) by Gender")
plt.xlabel("Gender")
plt.ylabel("Average Final Grade (G3)")
plt.grid(axis='y')
plt.show()
```
